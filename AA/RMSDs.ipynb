```python
structure="../Trajectories_and_tprs/PMB1s.tpr"
trajectory="../Trajectories_and_tprs/PMB1s_whole_nojump.xtc"
folder="../RMSDs/"
```

```python
RMSD_list=[]


for i,j in zip(["1:188","189:376","565:752","941:1128","1129:1316","1317:1504","1693:1880","1881:2068","2633:2820","2821:3008","3197:3384","3385:3572","3573:3760"],["1457_1644","1645_1832","2021_2208","2397_2584","2585_2772","2773_2960","3149_3336","3337_3524","4089_4276","4277_4464","4653_4840","4841_5028","5029_5216"]):
    RMSD_obj=rmsd_calc.RMSD_calculator(tpr=structure,traj=trajectory,saving_folder=folder,sel=f"resname PMB and (resid {i})",selname=f"PMB1_{j}")
    print(RMSD_obj)
    RMSD=RMSD_obj.RMSD_calculation()
    RMSD_list.append(RMSD)
    
```

```python
fig,ax=plt.subplots(ncols=1,nrows=1,figsize=(5,5))
for i in range(0,13):
    ax.plot(RMSD_list[i][1,:]/1000,RMSD_list[i][2,:])
                 
ax.set_xlabel("Time (ns)")
ax.set_ylabel(r"RMSD ($\AA$)")

```
