## Step 1

- Set up input arguments for rmsd_calc.RMSD_calculator class.

```python
sys.path.insert(0,os.path.abspath('../../Analysis_scripts'))
import rmsd_calc
```

```python
folder="../RMSDs/" #folder to save outputs into
protein_structure="../Trajectories_and_tprs/Proteins.tpr" #tpr filepath
protein_trajectory="../Trajectories_and_tprs/Proteins_whole.xtc" 
#trajectory filepath - proteins should be made whole using gmx trjconv

```

## Step 2 
- Create lists containing:
    1. Atom selections for individual protein monomers,
    2. Labels of individual protein monomers.

- Iterate through each protein monomer in residues_list and names_list:
  - Initialise RMSD_calculator class and run RMSD_calculation() function from t = 0 ns onwards.
  - Protein RMSD files are saved automatically by RMSD_calculation().



```python
residues_list=["1:417","418:834","835:1251"] # list containing atom selections for protein monomers
names_list=["Protein1","Protein2","Protein3"] # list containing labels for protein monomers

for i,j in zip(residues_list,names_list):
    RMSD_protein_obj=rmsd_calc.RMSD_calculator(tpr=protein_structure,traj=protein_trajectory,saving_folder=folder,sel=f"protein and (resid {i})",selname=f"{j}",sel2="backbone")
    RMSD_protein=RMSD_protein_obj.RMSD_calculation()
    
```

## Step 3
- Save tpr and trajectory for each individual protein and make whole (gmx trjconv).



## Step 4
- Iterate through each protein monomer in names_list:
  - Initialise RMSF_calculator class and run RMSF_calculation() function from t = 400 ns onwards for each protein monomer.
  - Protein RMSF files are saved automatically by RMSF_calculation().



```python
for i in names_list:
    RMSF=rmsd_calc.RMSF_calculator(tpr=f"{i}.gro",traj=f"{i}.xtc",saving_folder=folder,selname=f"{i}",ref_frame=40000,start=40000)
    RMSF_results=RMSF.RMSF_calculation()
```
