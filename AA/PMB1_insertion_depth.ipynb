```python
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import sys
import os

%matplotlib inline
```

```python
sys.path.insert(0,os.path.abspath('../../Analysis_scripts'))
import pmb1_com as com  
```

```python
sns.set_style('white')
sns.set_style("ticks")
plt.rcParams['font.size']='16'
sns.color_palette()
```

```python
structure="../Trajectories_and_tprs/No_solvent_CAs.tpr"
trajectory="../Trajectories_and_tprs/No_jump.xtc"
folder="../COM_analysis/"

Phosphates="(resname RAMP and name PA PB) or (resname POPE POPG and name P*)"
PMB1_sel='resname PMB'
PMB1_tail_sel="resname PMB and name C1 C2 C3 C4 C5 C6 C7 C8"
PMB1_Leu_sel="resname PMB and name C39 C40 C41 C42"
PMB1_Phe_sel='resname PMB and name C46 C47 C48 C49 C50 C51 C45'
PMB1_Dab_sel='resname PMB and name N9 N11 N16 N2 N5'
```

```python
PMB1_upper_whole=com.COM(tpr=structure,traj=trajectory,saving_folder=folder,sel=PMB1_sel,selname="PMB1_upper_whole",phos_dat=True,phosphates=Phosphates,step=5)
Time,Phosphates,PMB1_upper_whole_total=PMB1_upper_whole.COM_calculator()
```

```python
PMB1_tail_upper=com.COM(tpr=structure,traj=trajectory,saving_folder=folder,sel=PMB1_tail_sel,selname="PMB1_tail_upper",phosphates=Phosphates,step=5)
Time,PMB1_tail_upper_total=PMB1_tail_upper.COM_calculator()
```

```python
PMB1_Leu_upper=com.COM(tpr=structure,traj=trajectory,saving_folder=folder,sel=PMB1_Leu_sel,selname="PMB1_Leu_upper",phosphates=Phosphates,step=5)
Time,PMB1_Leu_upper_total=PMB1_Leu_upper.COM_calculator()
```

```python
PMB1_Phe_upper=com.COM(tpr=structure,traj=trajectory,saving_folder=folder,sel=PMB1_Phe_sel,selname="PMB1_Phe_upper",phosphates=Phosphates,step=5)
Time,PMB1_Phe_upper_total=PMB1_Phe_upper.COM_calculator()
```

```python
PMB1_Dab_upper=com.COM(tpr=structure,traj=trajectory,saving_folder=folder,sel=PMB1_Dab_sel,selname="PMB1_Dab_upper",phosphates=Phosphates,step=5)
Time,PMB1_Dab_upper_total=PMB1_Dab_upper.COM_calculator()
```
